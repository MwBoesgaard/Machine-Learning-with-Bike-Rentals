## Introduction

<font size="3"> Many cities have a communal bike sharing system for renting bicycles for a small fee. Washington, D.C, USA, has such a system and has been collecting detailed data on the number of bicycle rentals. Hadi Fanaee-T at the University of Porto compiled this data into a CSV file. The file contains 17380 rows, with each row representing the number of bike rentals for a single hour of a single day.

In this machine learning project we will attempt to predict the number of bikers for a given hour of the day based on the time and the weather. Let's dive in by reading the data into pandas and having a closer look!</font>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

bike_rentals = pd.read_csv('bike_rental_hour.csv')
bike_rentals.info()
bike_rentals.head()
```

<font size="3">
We can see the dataset contains 17379 rows, with no rows missing for any of the attributes. 
    
Except for the dteday column, which contains a datetime, all other columns are numerical (ints or floats).
 
Since the dataset has no missing values, no string columns and contains several normalized attributes (representeted as a decimal point value), such as **'temp'**, **'hum'** and **'windspeed'** it is safe to assume this dataset has already been thoroughly cleaned and feature engineered in preparation for use in Machine Learning models...

Before we continue a detailed description of all the columns in the dataset can be found below.
    
In this project we are interested in predicting the number of bike rentals per hour, i.e. the **'cnt'** column.    


</font>


#### Column Descriptions
- **instant** - A unique sequential ID number for each row

- **dteday** - The date of the rentals

- **season** - The season in which the rentals occurred

- **yr** - The year the rentals occurred

- **mnth** - The month the rentals occurred

- **hr** - The hour the rentals occurred

- **holiday** - Whether or not the day was a holiday

- **weekday** - The day of the week (as a number, 0 to 7)

- **workingday** - Whether or not the day was a working day

- **weathersit** - The weather (as a categorical variable)

- **temp** - The temperature, on a 0-1 scale

- **atemp** - The adjusted temperature

- **hum** - The humidity, on a 0-1 scale

- **windspeed** - The wind speed, on a 0-1 scale

- **casual** - The number of casual riders (people who has not signed up with the bike sharing program)

- **registered** - The number of registered riders

- **cnt** - The total number of bike rentals (casual + registered)

<font size="3">
Despite the datatypes for the columns being listed as ints and floats, <u>several of them are categorical in nature.</u> 
    
To see which, let us find the number of unique values for each column

</font>


```python
def unique_values_per_col(df):
    results = pd.DataFrame()
    for col in df.columns:
        count = bike_rentals[col].value_counts().index.value_counts().sum()
        results.loc[0,col] = count
    return results.head(5)
    
unique_values_per_col(bike_rentals)
```

<font size="3">
    The following columns have only a few unique values (but more than 2) and upon inspection can be seen to contain categorical data:
    
- season
- yr
- mnth
- hr
- weathersit

Catagorical data types are not a problem for tree-based machine learning models (such as decision trees and random forests), but can be a problem for linear models (such as linear regression). In this project we will utilize both types of model and therefore it might be useful to transform the categorical data using [One-Hot Encoding](https://en.wikipedia.org/wiki/One-hot) (splitting each category into its own column). But before we do any feature-engineering of our own, let us have a closer look at our target label, **'cnt'**.
</font>

```python
fig, (ax1) = plt.subplots(figsize = (16,8))


ax1.hist(bike_rentals['cnt'], bins=20)
ax1.set_xticks(range(0,1001,50))
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.set_xlabel("Bikers per hour", fontsize = '18')
ax1.set_ylabel("count", fontsize = '18')
ax1.tick_params(axis = 'x', labelsize=14, rotation = 45)
ax1.tick_params(axis = 'y', labelsize=14)
ax1.patches[0].set_color('teal')

plt.annotate("",
              xy=(60, 4800), xycoords='data',
              xytext=(420, 4850), textcoords='data',
              arrowprops=dict(arrowstyle="-", linestyle = '--',
                              connectionstyle="arc3,rad=0."), 
              )

plt.annotate("",
              xy=(60, 4700), xycoords='data',
              xytext=(420, 2200), textcoords='data',
              arrowprops=dict(arrowstyle="-", linestyle = '--',
                              connectionstyle="arc3,rad=0."), 
              )

ax2 = plt.axes([0.5, 0.45, 0.4, 0.4])
ax2.hist(bike_rentals['cnt'][bike_rentals['cnt']<=50], bins=10)
ax2.tick_params(axis = 'x', labelsize=8)
ax2.tick_params(axis = 'y', labelsize=8)
ax2.set_xticks(range(0,51,5))
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.set_xlabel("Bikers per hour", fontsize = '12')
ax2.set_ylabel("count", fontsize = '12')
ax2.tick_params(labelsize=14)

for i in range(0,10):
    ax2.patches[i].set_color('teal')


plt.show()
```

<font size="3">Looking at the histogram of all the **'cnt'** data points above, it is clear that it does not follow a [Gaussian Distribution](https://en.wikipedia.org/wiki/Normal_distribution), but is heavily skewed to the right. This is a problem for our future modelling as many of the techniques used assume that the data follows a gaussian distribution. To remedy this we can transform our target label using a log transformation (as seen below).</font>

```python
fig, (ax1) = plt.subplots(figsize = (16,8))


ax1.hist(np.log(bike_rentals['cnt']), bins=20)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.set_xlabel("log (Bikers per hour)", fontsize = '18')
ax1.set_ylabel("count", fontsize = '18')
ax1.tick_params(axis = 'x', labelsize=14, rotation = 45)
ax1.tick_params(axis = 'y', labelsize=14)


plt.show()
```

<font size="3">Although the log tranformation did not manage to make the **'cnt'** data follow a normal distribution perfectly, it is still much better. Let us add this log-transformed label to our dataset, and use it as our target going forward. 
    
We can always use transform the data back to actual riders when needed. </font>

```python
bike_rentals['log_cnt'] = np.log(bike_rentals['cnt'])
```

To start with the feature engineering, let us first make a seperate data-set for the linear model and transform the catagorical features as mentioned...</font>

```python
bike_rentals_lin = bike_rentals.copy()
```

```python
one_hot = pd.get_dummies(bike_rentals_lin['hr'], drop_first = True, prefix = 'hr_')
bike_rentals_lin = bike_rentals_lin.drop( 'hr', axis = 1)
bike_rentals_lin = bike_rentals_lin.join(one_hot)
bike_rentals_lin.head()
```

```python
def assign_label_mnth(month):
    if (month == 1):
        return "january"
    elif (month == 2):
        return "february"
    elif (month == 3):
        return "march"
    elif (month == 4):
        return "april"
    elif (month == 5):
        return "may"
    elif (month == 6):
        return "june"
    elif (month == 7):
        return "july"
    elif (month == 8):
        return "august"
    if (month == 9):
        return "september"
    elif (month == 10):
        return "october"
    elif (month == 11):
        return "november"
    else:
        return "december"
    
bike_rentals_lin['mnth_label'] = bike_rentals_lin['mnth'].map(assign_label_mnth)
one_hot = pd.get_dummies(bike_rentals_lin['mnth_label'], drop_first = True)
bike_rentals_lin = bike_rentals_lin.drop(['mnth_label', 'mnth', 'season'], axis = 1)
bike_rentals_lin = bike_rentals_lin.join(one_hot)
bike_rentals_lin.head()
```

```python
def assign_label_bad_weather(weather):
    if (weather == 4):
        return "horrible weather"
    elif (weather == 3):
        return "bad weather"
    elif (weather == 2):
        return "fair weather"
    else:
        return "perfect weather"
    
bike_rentals_lin['weather_label'] = bike_rentals_lin['weathersit'].map(assign_label_bad_weather)
one_hot = pd.get_dummies(bike_rentals_lin['weather_label'], drop_first = True)
bike_rentals_lin = bike_rentals_lin.drop(['weather_label', 'weathersit'], axis = 1)
bike_rentals_lin = bike_rentals_lin.join(one_hot)
bike_rentals_lin.head()
```

```python
training = bike_rentals.sample((int(bike_rentals.shape[0]*0.8)))
print("training pct of dataset: " + str(training.shape[0]/bike_rentals.shape[0]*100))
test = pd.merge(training, bike_rentals, how='outer', indicator=True)
test = test[test['_merge'] == 'right_only'].copy()
test.drop('_merge', axis=1)
print("test pct of dataset: " + str(test.shape[0]/bike_rentals.shape[0]*100))

training_lin = bike_rentals_lin.sample((int(bike_rentals_lin.shape[0]*0.8)))

test_lin = pd.merge(training_lin, bike_rentals_lin, how='outer', indicator=True)
test_lin = test_lin[test_lin['_merge'] == 'right_only'].copy()
test_lin.drop('_merge', axis=1)
```

```python
#Since the linear model can extrapolate better than the tree models
#... we can improve the fit by removing outliers (Z-scores more than 3)

from scipy import stats
training_lin['cnt_z_score'] = stats.zscore(training_lin['cnt'])
num_of_outliers = training_lin[training_lin['cnt_z_score'] > 3].shape[0]
print("Number of outliers: " + str(num_of_outliers))

lin_training = training_lin[training_lin['cnt_z_score'] < 3]
```

<font size="3">The datasets are not prepared for our Machine Learning models, first let us define our feature columns and target column.

Naturally we will want to remove columns that our target column is derived from i.e. **'casual'**, **'registered'**, as well as the redundant/unusable columns **'instant'** and **'dteday'**, as well as **'cnt_z_score'** for the linear model. We will drop **'atemp'** too, since we already have a **'temp'** column</font>

```python
features = training.columns.tolist()
features_lin = training_lin.columns.tolist()
list_of_features_to_drop = ['cnt', 'casual', 'dteday', 'registered', 'log_cnt', 'instant', 'cnt_z_score', 'atemp', 'workingday']
features = [x for x in features if (x not in list_of_features_to_drop)]
features_lin = [x for x in features_lin if (x not in list_of_features_to_drop)]
print("Feature cols: (tree model) " + str(features))
print("No. of features: " + str(len(features)) + '\n')

print("Feature cols (linear model): " + str(features_lin))
print("No. of features: " + str(len(features_lin)) + '\n')

target_col = 'log_cnt'

print("Target col: " + str(target_col))
```

```python
corr = bike_rentals.corr()
cnt_corr = corr['log_cnt'].sort_values()
cnt_corr = cnt_corr.drop(['cnt', 'registered', 'casual', 'instant', 'log_cnt', 'atemp', 'workingday']) # Let's remove the target column and those directly derived from it

f, ax = plt.subplots(figsize=(16, 7))
ax = sns.barplot(cnt_corr.index, cnt_corr)
ax.set_title("Tree optimized model", fontsize = '18')
ax.set_ylabel("log cnt", fontsize = '18')
ax.tick_params(axis = 'x', labelsize = 20, rotation = 90)
ax.tick_params(axis = 'y', labelsize = 16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

```python
corr_lin = bike_rentals_lin.corr()
cnt_corr_lin = corr_lin['log_cnt'].sort_values()
cnt_corr_lin = cnt_corr_lin.drop(['cnt', 'registered', 'casual', 'instant', 'log_cnt', 'atemp', 'workingday']) # Let's remove the target column and those directly derived from it

f, ax = plt.subplots(figsize=(16, 7))
ax = sns.barplot(cnt_corr_lin.index, cnt_corr_lin)
ax.set_title("Linear optimized model", fontsize = '18')
ax.set_ylabel("log cnt", fontsize = '18')
ax.tick_params(axis = 'x', labelsize = 20, rotation = 90)
ax.tick_params(axis = 'y', labelsize = 16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

## Model Training, Fit and Predictions
<font size="3">Let's start by building a linear regression model using the sklearn library and see if our dataset optimized for the linear model will do better.</font>

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.compose import TransformedTargetRegressor

def linear_model_regression(training, test, features):
    lr = LinearRegression()

    lr.fit(training[features], training[target_col])

    training_predictions = lr.predict(training[features])
    training_mae = mean_absolute_error(np.exp(training[target_col]), np.exp(training_predictions))
    training_mse = mean_squared_error(np.exp(training[target_col]), np.exp(training_predictions))
    training_rmse = training_mse**(1/2)

    test_predictions = lr.predict(test[features])
    test_mae = mean_absolute_error(np.exp(test[target_col]), np.exp(test_predictions))
    test_mse = mean_squared_error(np.exp(test[target_col]), np.exp(test_predictions))
    test_rmse = test_mse**(1/2)

    print("training_mae: " + str(training_mae))
    print("training_mse: " + str(training_mse))
    print("training_rmse: " + str(training_rmse))
    print("")
    print("test_mae: " + str(test_mae))
    print("test_mse: " + str(test_mse))
    print("test_rmse: " + str(test_rmse))
    print("")
    
    return test_predictions

print("Tree optimized dataset:")
old_model = linear_model_regression(training, test, features)
print("Linear optimized dataset:")
new_model = linear_model_regression(training_lin, test_lin, features_lin)

comparison = test.copy()
comparison_lin = test_lin.copy()
comparison['cnt_guess_linear'] = np.exp(old_model)
comparison_lin['cnt_guess_linear_new'] = np.exp(new_model)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))

ax1.scatter(comparison['cnt'], comparison['cnt_guess_linear'])
ax1.set_title("Tree optimized dataset", fontsize = '18')
ax1.set_xlabel("Actual cnt", fontsize = '18')
ax1.set_ylabel("Predicted cnt", fontsize = '18')
ax1.tick_params(labelsize=14)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)

ax2.scatter(comparison_lin['cnt'], comparison_lin['cnt_guess_linear_new'])
ax2.set_title("Linear optimized dataset", fontsize = '18')
ax2.set_xlabel("Actual cnt", fontsize = '18')
ax2.set_ylabel("Predicted cnt", fontsize = '18')
ax2.tick_params(labelsize=14)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)

plt.show
```

<font size="3"> Awesome! We can see that by transforming the catagorical features with One-Hot Encoding, managed to reduce the test RMSE of the linear model from 163.4 (Tree optimized dataset) to 101.0 (Linear optimized dataset)!
 
Next, let's build a decision tree model and do some hyperparameter optimization to find the parameters that will yield the best model.</font>

```python
from sklearn.tree import DecisionTreeRegressor

def decision_tree_optimize_1(test_dataframe, msl, md):
    dtr = DecisionTreeRegressor(min_samples_leaf = msl, max_depth = md)
    
    dtr.fit(training[features], training[target_col])

    predictions = dtr.predict(test_dataframe[features])
    mae = mean_absolute_error(np.exp(test_dataframe[target_col]), np.exp(predictions))

    return mae

def decision_tree_optimize_2(test_dataframe, mss, mf):
    dtr = DecisionTreeRegressor(min_samples_split = mss, max_features = mf)
    
    dtr.fit(training[features], training[target_col])

    predictions = dtr.predict(test_dataframe[features])
    mae = mean_absolute_error(np.exp(test_dataframe[target_col]), np.exp(predictions))

    return mae

def plot_two_features(outer_feature_name, inner_feature_name, outer_feature_list, inner_feature_list, training_ax, test_ax, decision_tree_function):
    for index, i in enumerate(outer_feature_list):
        x_vals = []
        training_vals = []
        test_vals = []
        #colors = ['orange', 'green', 'red', 'blue', 'purple', 'yellow', 'black']

        for j in inner_feature_list:
                x_vals.append(j)
                training_vals.append(decision_tree_function(training, j, i))
                test_vals.append(decision_tree_function(test, j, i))

        training_ax.scatter(x_vals, training_vals, label = outer_feature_name+": "+ str(i), cmap=plt.get_cmap('Wistia'))
        training_ax.legend(loc='center left', bbox_to_anchor=(-0.5, 0.5))
        training_ax.set_xlabel(inner_feature_name, fontsize = 16.0)
        training_ax.set_ylabel("mean average error", fontsize = 16.0)
        training_ax.tick_params(labelsize=14)
        training_ax.set_title("Training Data", fontsize = 24.0)
        training_ax.spines['right'].set_visible(False)
        training_ax.spines['top'].set_visible(False)
        
        
        test_ax.scatter(x_vals, test_vals, label = outer_feature_name+": "+ str(i), cmap=plt.get_cmap('Wistia'))
        test_ax.set_xlabel(inner_feature_name, fontsize = 16.0)
        test_ax.set_ylabel("mean average error", fontsize = 16.0)
        test_ax.tick_params(labelsize=14)
        #test_ax.set_ylim(0,75)
        test_ax.set_title("Test Data", fontsize = 24.0)
        test_ax.spines['right'].set_visible(False)
        test_ax.spines['top'].set_visible(False)

min_samples_leaf = range(1,50)
max_depth_range = range(5,30,5)
min_samples_split = range(2,60)
max_features = range(5,len(features)+1,1)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize=(16,16))
fig.tight_layout(pad=8.0)

plot_two_features("max_depth", "min_samples_leaf", max_depth_range, min_samples_leaf, ax1, ax2, decision_tree_optimize_1)
plot_two_features("max_features", "min_samples_split", max_features, min_samples_split, ax3, ax4, decision_tree_optimize_2)
    
plt.show()
```

```python
def decision_regression(training, test, features):

    dtr = DecisionTreeRegressor(min_samples_leaf = 10, max_depth = 25, min_samples_split = 15)

    dtr.fit(training[features], training[target_col])

    training_predictions = dtr.predict(training[features])
    training_mae = mean_absolute_error(np.exp(training[target_col]), np.exp(training_predictions))
    training_mse = mean_squared_error(np.exp(training[target_col]), np.exp(training_predictions))
    training_rmse = training_mse**(1/2)

    test_predictions = dtr.predict(test[features])
    test_mae = mean_absolute_error(np.exp(test[target_col]), np.exp(test_predictions))
    test_mse = mean_squared_error(np.exp(test[target_col]), np.exp(test_predictions))
    test_rmse = training_mse**(1/2)


    print("training_mae: " + str(training_mae))
    print("training_mse: " + str(training_mse))
    print("training_rmse: " + str(training_rmse))
    print("")
    print("test_mae: " + str(test_mae))
    print("test_mse: " + str(test_mse))
    print("test_rmse: " + str(test_rmse))
    print("")
    
    return test_predictions

print("Tree optimized dataset:")
old_model = decision_regression(training, test, features)
print("Linear optimized dataset:")
new_model = decision_regression(training_lin, test_lin, features_lin)

comparison['cnt_guess_tree'] = np.exp(old_model)
comparison_lin['cnt_guess_tree_new'] = np.exp(new_model)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))

ax1.scatter(comparison['cnt'], comparison['cnt_guess_tree'])
ax1.set_title("Tree optimized dataset", fontsize = '18')
ax1.set_xlabel("Actual cnt", fontsize = '18')
ax1.set_ylabel("Predicted cnt", fontsize = '18')
ax1.tick_params(labelsize=14)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)

ax2.scatter(comparison_lin['cnt'], comparison_lin['cnt_guess_tree_new'])
ax2.set_title("Linear optimized dataset", fontsize = '18')
ax2.set_xlabel("Actual cnt", fontsize = '18')
ax2.set_ylabel("Predicted cnt", fontsize = '18')
ax2.tick_params(labelsize=14)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)

plt.show
```

```python
from sklearn.ensemble import RandomForestRegressor

def random_forest_optimize_1(test_dataframe, msl, md):
    rfr = RandomForestRegressor(min_samples_leaf = msl, max_depth = md)
    rfr.fit(training[features], training[target_col])

    predictions = rfr.predict(test_dataframe[features])
    mae = mean_absolute_error(np.exp(test_dataframe[target_col]), np.exp(predictions))

    return mae

def random_forest_optimize_2(test_dataframe, mss, mf):
    rfr = RandomForestRegressor(min_samples_split = mss, n_estimators = mf)
    rfr.fit(training[features], training[target_col])

    predictions = rfr.predict(test_dataframe[features])
    mae = mean_absolute_error(np.exp(test_dataframe[target_col]), np.exp(predictions))

    return mae

min_samples_leaf = range(1,50,10)
max_depth_range = range(5,55,10)
min_samples_split = range(2,100,10)
n_estimators = range(100,201,100)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize=(16,16))
fig.tight_layout(pad=8.0)

plot_two_features("max_depth", "min_samples_leaf", max_depth_range, min_samples_leaf, ax1, ax2, random_forest_optimize_1)
plot_two_features("n_estimators", "min_samples_split", n_estimators, min_samples_split, ax3, ax4, random_forest_optimize_2)
    
plt.show()
```

```python
def random_forest_regression(training, test, features):

    rfr = RandomForestRegressor(min_samples_leaf = 10, max_depth = 25, min_samples_split = 20)

    rfr.fit(training[features], training[target_col])

    training_predictions = rfr.predict(training[features])
    training_mae = mean_absolute_error(np.exp(training[target_col]), np.exp(training_predictions))
    training_mse = mean_squared_error(np.exp(training[target_col]), np.exp(training_predictions))
    training_rmse = training_mse**(1/2)

    test_predictions = rfr.predict(test[features])
    test_mae = mean_absolute_error(np.exp(test[target_col]), np.exp(test_predictions))
    test_mse = mean_squared_error(np.exp(test[target_col]), np.exp(test_predictions))
    test_rmse = test_mse**(1/2)
    
    print("training_mae: " + str(training_mae))
    print("training_mse: " + str(training_mse))
    print("training_rmse: " + str(training_rmse))
    print("")
    print("test_mae: " + str(test_mae))
    print("test_mse: " + str(test_mse))
    print("test_rmse: " + str(test_rmse))
    print("")
    
    return test_predictions

print("Tree optimized dataset:")
old_model = random_forest_regression(training, test, features)
print("Linear optimized dataset:")
new_model = random_forest_regression(training_lin, test_lin, features_lin)

comparison['cnt_guess_forest'] = np.exp(old_model)
comparison_lin['cnt_guess_forest_new'] = np.exp(new_model)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))

ax1.scatter(comparison['cnt'], comparison['cnt_guess_forest'])
ax1.set_title("Tree optimized dataset", fontsize = '18')
ax1.set_xlabel("Actual cnt", fontsize = '18')
ax1.set_ylabel("Predicted cnt", fontsize = '18')
ax1.tick_params(labelsize=14)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)

ax2.scatter(comparison_lin['cnt'], comparison_lin['cnt_guess_forest_new'])
ax2.set_title("Linear optimized dataset", fontsize = '18')
ax2.set_xlabel("Actual cnt", fontsize = '18')
ax2.set_ylabel("Predicted cnt", fontsize = '18')
ax2.tick_params(labelsize=14)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)

plt.show
```
